```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_excel("dataset2.xlsx", index_col=0)
dataset
```

```python
# 3) 변환된 데이터 타입 확인
print(dataset.dtypes)
```

```python
### 0. ADF 단위근 검정

from statsmodels.tsa.stattools import adfuller

# 결과 저장용 리스트
results = []

# DataFrame의 각 열에 대해 반복
for col in dataset.columns:
    ADF_result = adfuller(dataset[col], maxlag=1)
    pd.set_option('display.float_format', '{:.6f}'.format)
    # 결과 저장 (열 이름, 통계량, p값, 유의 수준 임계값들)
    results.append({
        'Column': col,
        'ADF Statistic': ADF_result[0],
        'p-value': ADF_result[1],
        '1% Critical Value': ADF_result[4]['1%'],
        '5% Critical Value': ADF_result[4]['5%'],
        '10% Critical Value': ADF_result[4]['10%'],
        'Is Stationary (p<0.05)': ADF_result[1] < 0.05
    })

# DataFrame으로 변환
results = pd.DataFrame(results)
results
```

```python
import math
import matplotlib.pyplot as plt
import matplotlib as mpl

years = dataset.index

mpl.rcParams['font.family'] = 'Malgun Gothic'   # Windows: Malgun Gothic
mpl.rcParams['axes.unicode_minus'] = False      # 마이너스 기호 깨짐 방지

n_vars = len(dataset.columns)
n_cols = 4
n_rows = math.ceil(n_vars / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(50, 6*n_rows), sharex=True)
axes = axes.flatten()

for ax, col in zip(axes, dataset.columns):
    ax.plot(years, dataset[col], marker='o')
    ax.set_title(col, fontsize=10)
    ax.tick_params(axis='x', rotation=45)

# 남은 빈 플롯 숨기기
for ax in axes[n_vars:]:
    ax.set_visible(False)

plt.tight_layout()
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
import math

# ACF & PACF를 모든 변수에 대해 시각화
n_vars = len(dataset.columns)
fig, axes = plt.subplots(n_vars, 2, figsize=(14, 4 * n_vars))

for i, col in enumerate(dataset.columns):
    # ACF
    plot_acf(dataset[col].dropna(), ax=axes[i, 0], lags=20, zero=False)
    axes[i, 0].set_title(f"{col} - ACF")

    # PACF
    plot_pacf(dataset[col].dropna(), ax=axes[i, 1], lags=20, zero=False, method='ywm')
    axes[i, 1].set_title(f"{col} - PACF")

plt.tight_layout()
plt.show()
```

```python
# 사용자 변환 지정
transform_map = {
    '소비자물가총지수': 'log',
    '매출액': 'log',
    '수출': 'log',
    '연구개발비': 'log',
    '소비자심리지수': 'boxcox',
    '실업률': 'boxcox',
    '가동률지수': 'boxcox',
    '환율': 'log',
    '기준금리': 'sqrt',
    'ROE' : 'log'
}
```

```python
## 정상성 확보를 위한 변환

import numpy as np
import pandas as pd
from scipy.stats import boxcox

df_adf = dataset.copy()
bc_lambdas = {}

for col, method in transform_map.items():
    s = dataset[col]

    # (1) 결측/0/음수 처리
    if method == 'boxcox':
        # 모두 양수인지 확인
        if (s <= 0).any():
            shift = abs(s.min()) + 1e-6
            s = s + shift
        vals, lam = boxcox(s.dropna())
        bc_lambdas[col] = lam
        # 인덱스 매칭해서 할당
        df_adf[col] = pd.Series(vals, index=s.dropna().index)

    elif method == 'log':
        df_adf[col] = np.log(s.replace(0, np.nan)).fillna(method='bfill')

    elif method == 'sqrt':
        df_adf[col] = np.sqrt(s.clip(lower=0))

    else:  # 원본 유지
        df_adf[col] = s

# 상수열 제거
consts = [c for c in df_adf.columns if df_adf[c].nunique() <= 1]
if consts:
    df_adf.drop(columns=consts, inplace=True)
    print("상수열 제거:", consts)

# 결과 출력
pd.set_option('display.max_columns', None)
print("=== ADF용 변환 데이터 ===")
df_adf
```

```python
### 1. ADF 단위근 검정

from statsmodels.tsa.stattools import adfuller

# 결과 저장용 리스트
results = []

# DataFrame의 각 열에 대해 반복
for col in df_adf.columns:

    ADF_result = adfuller(df_adf[col], maxlag=1)
    pd.set_option('display.float_format', '{:.6f}'.format)
    # 결과 저장 (열 이름, 통계량, p값, 유의 수준 임계값들)
    results.append({
        'Column': col,
        'ADF Statistic': ADF_result[0],
        'p-value': ADF_result[1],
        '1% Critical Value': ADF_result[4]['1%'],
        '5% Critical Value': ADF_result[4]['5%'],
        '10% Critical Value': ADF_result[4]['10%'],
        'Is Stationary (p<0.05)': ADF_result[1] < 0.05
    })

# 보기 좋게 DataFrame으로 변환
results = pd.DataFrame(results)
results
```

```python
## 1차 차분

# (1) 비정상 변수 리스트
non_stat = results.loc[results['Is Stationary (p<0.05)']==False, 'Column'].tolist()

# (2) df_adf 복사
df_diff1 = df_adf.copy()

# (3) 비정상 변수에만 diff 적용
df_diff1[non_stat] = df_diff1[non_stat].diff()

# (4) 첫 번째 행의 NaN 제거 (선택)
df_diff1 = df_diff1.iloc[1:]

# (5) 결과 확인
print("1차 차분을 적용한 변수:", non_stat)
pd.set_option('display.max_columns', None)
print("=== df_adf 전체 중, 비정상 변수만 1차 차분 적용 결과 ===")
df_diff1
```

```python
### 2. ADF 단위근 검정

from statsmodels.tsa.stattools import adfuller

# 결과 저장용 리스트
results2 = []

# DataFrame의 각 열에 대해 반복
for col in df_diff1.columns:
    ADF_result = adfuller(df_diff1[col], maxlag=1)
    pd.set_option('display.float_format', '{:.6f}'.format)
    # 결과 저장 (열 이름, 통계량, p값, 유의 수준 임계값들)
    results2.append({
        'Column': col,
        'ADF Statistic': ADF_result[0],
        'p-value': ADF_result[1],
        '1% Critical Value': ADF_result[4]['1%'],
        '5% Critical Value': ADF_result[4]['5%'],
        '10% Critical Value': ADF_result[4]['10%'],
        'Is Stationary (p<0.05)': ADF_result[1] < 0.05
    })

# 보기 좋게 DataFrame으로 변환
results2 = pd.DataFrame(results2)
results2
```
